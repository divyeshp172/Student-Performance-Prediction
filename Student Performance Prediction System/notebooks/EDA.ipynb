# 🎓 Exploratory Data Analysis (EDA)
This notebook performs EDA on the Student Performance dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('../data/student_performance.csv', sep=None, engine='python')
df.head()
```

## 1. Dataset Inspection

```python
print("Shape:", df.shape)
print("\n--- Info ---")
df.info()
print("\n--- Describe ---")
display(df.describe())
print("\n--- Missing Values ---")
display(df.isnull().sum())
print("\n--- Duplicate Values ---")
print("Duplicates:", df.duplicated().sum())
```

## 2. Univariate Analysis

```python
# Histograms for numerical columns
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
# Boxplots to check outliers
for col in num_cols:
    plt.figure(figsize=(8, 2))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

## 3. Bivariate & Multivariate Analysis

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot if dataset size allows
if len(num_cols) <= 10:
    sns.pairplot(df[num_cols])
    plt.show()
else:
    print("Too many numerical columns for a clear pairplot.")
```
